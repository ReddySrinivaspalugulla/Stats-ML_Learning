```python
import numpy as np
```

```python
def gradient_descent(x,y):
    m = 0  # initial value of slope
    c = 0  # initial value of intercept
    learning_rate = 0.01  # initial value of alpha
    itr = 10000  # number of steps to move on the error function
    n = len(x)  # number of rows in data
    
    for i in range(itr):
        
        y_pred = m * x + c
        
        cost = (1/n) * sum([val**2 for val in (y - y_pred)])
        
        dm = (-2/n) * sum(x * (y - y_pred))
        
        dc = (-2/n) * sum(y - y_pred)
        
        m = m - learning_rate * dm
        
        c = c - learning_rate * dc
        
        print("Slope = {} | Intercept = {} | Itr = {} | Cost = {}".format(m,c,i+1,cost))
    
    
```

```python
x = np.array([1,2,3,4,5])
y = np.array([19.88,64.88,109.88,154.88,199.88])
```

```python
gradient_descent(x,y)
```
